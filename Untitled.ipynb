```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data
df = pd.read_csv('listings.csv')

# Data cleaning
df = df.drop(['id', 'name', 'host_id', 'host_name', 'reviews_per_month','last_review', 'license','neighbourhood_group','minimum_nights','number_of_reviews','calculated_host_listings_count','availability_365','number_of_reviews_ltm'], axis=1)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['price'] >= Q1 - 1.5*IQR) & (df['price'] <= Q3 + 1.5*IQR)]
```

```python
# Data visualization
plt.figure(figsize=(12,8))
sns.histplot(data=df, x="price", bins=50, kde=True)
plt.title("Distribution of Airbnb Prices")
plt.show()

plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="longitude", y="latitude", hue="price", palette="coolwarm")
plt.title("Price by Location")
plt.show()

```

```python
# One-hot encode the room_type column
room_type_dummies = pd.get_dummies(df['room_type'], prefix='room_type')

# Concatenate the one-hot encoded columns with the original DataFrame
df = pd.concat([df, room_type_dummies], axis=1)

# Drop the original room_type column
df.drop('room_type', axis=1, inplace=True)
df
```

```python
# Feature engineering
X = df.drop('price', axis=1)
y = df['price']
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam

# Define the model architecture
model = Sequential([
    Dense(32, input_shape=(X_train.shape[1],)),
    Activation('relu'),
    Dense(16),
    Activation('relu'),
    Dense(1)
])

# Compile the model
model.compile(loss='mean_absolute_error', optimizer=Adam(lr=0.001))

# Print a summary of the model architecture
model.summary()
```

```python
# Train the model
history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=128)
```

```python
# Evaluate the model on the testing data
test_loss = model.evaluate(X_test, y_test)

# Print the test loss and accuracy
print(f'Test loss: {test_loss:.4f}')

# Plot the loss curve
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.legend()
plt.show()
predictions = model.predict(X_test)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, predictions)
print(mse)
accuracy = r2_score(y_test, predictions)
print(f'Accuracy: {accuracy:.4f}')
```

```python
# Fine-tune the model
model = Sequential([
    Dense(64, input_shape=(X_train.shape[1],)),
    Activation('relu'),
    Dense(32),
    Activation('relu'),
    Dense(16),
    Activation('relu'),
    Dense(1)
])

model.compile(loss='mean_absolute_error', optimizer=Adam(lr=0.0001))

history = model.fit(X_train, y_train, validation_split=0.2, epochs=200, batch_size=128)

# Evaluate the model on the testing data
test_loss = model.evaluate(X_test, y_test)

# Print the test loss and accuracy
print(f'Test loss: {test_loss:.4f}')
```

```python
# Plot the loss curve
plt.figure(figsize=(12,8))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.legend()
plt.show()
```
